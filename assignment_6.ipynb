```python
import selenium
import pandas as pd
from selenium import webdriver
```

```python
driver.get('https://www.flipkart.com/')
```

```python
search_sunglasses=driver.find_element_by_name('q')
search_sunglasses
```

```python
search_sunglasses.send_keys("sunglasses")
```

```python
search_btn=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_btn.click()
```

```python
Brand=[]
for page in range(0,3):
    for i in driver.find_elements_by_xpath('//div[@class="_2B099V"]/div[1]'):
        Brand.append(i.text)
    url1=driver.find_element_by_xpath("//a[@class='_1LKTO3']").get_attribute('href')
    driver.get(url1)
len(Brand)
```
