```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://bookpage.com/reviews")
page
```

```python
page.content
```

```python
Soup=BeautifulSoup(page.content)
Soup
```

```python
print(Soup.prettify())
```

```python
Book_name=[]

book_name=Soup.find_all('h4',class_="italic")

for i in book_name:

    Book_name.append(i.text.replace('\n',''))

Book_name
```

```python
Book_author=[]

book_author=Soup.find_all('p',class_="sans bold")

for i in book_author:

    Book_author.append(i.text.replace('\n',''))

Book_author
```

```python
Book_genre=[]

book_genre=Soup.find_all('p',class_="genre-links hidden-phone")

for i in book_genre:

    Book_genre.append(i.text.replace('\n',''))

Book_genre
```

```python
Book_review=[]

book_review=Soup.find_all('p',class_="excerpt")

for i in book_review:

    Book_review.append(i.text.replace('\n',''))

Book_review
```

```python
print(len(book_name),len(book_author),len(book_genre),len(book_review))
```

```python
import pandas as pd
books=pd.DataFrame({})
books['name']=Book_name
books['author']=Book_author
books['genre']=Book_genre
books['review']=Book_review
```
